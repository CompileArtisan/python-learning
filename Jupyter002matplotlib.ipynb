```python
import matplotlib.pyplot as plt
import numpy as np
```

## Some basic functions used

### <i> plt.plot() </i>
#### Syntax
        plt.plot([x], y, [fmt], data=None, **kwargs)
- x : array-like or scalar , meaning the x-axis values 
- y : array-like or scalar , meaning the y-axis values
- fmt : str, optional , meaning the format string, e.g. 'ro' for red circles, 'bs' for blue squares, 'g-' for green solid line
- data : indexable object, optional , meaning the data object ,  e.g. a pandas dataframe, a list, a numpy array

```python
# a simple plot
plt.plot([3, 2, 6, 4]) # y values
plt.ylabel('some numbers')
plt.show()
```

#### Plotting the equation y = x^2
        np.arange(start inclusive, end exclusive, step)

```python
x = np.arange(0., 5., 0.2) # x = [0, 0.2, 0.4, ..., 4.8]
y = x**2
plt.plot(x, y)
plt.show()
```

#### Plotting the equation y=sin(x)
        np.linspace(start included, end included, number of points)

```python
x = np.linspace(0, 10, 100) # x = [0, 0.1, 0.2, ..., 9.9, 10]
y = np.sin(x) 
plt.figure(figsize=(10, 6)) # width, height in inches
plt.plot(x, y, label='Sine Wave') # label for legend
plt.title('Basic Line Plot') # title
plt.xlabel('X-axis') 
plt.ylabel('Y-axis')
plt.legend() # show legend
plt.grid(True) # show grid
plt.show()
```

- arange() is used when you know start and end points, and <b> step value </b>
- linspace() is used when you know start and end points, and the <b> number of points </b>

#### Scatter Plots
        plt.scatter(x, y, s (colors), s (sizes), a (alpha), cmap (colormap))
- x, y : array_like, shape (n, )
The data positions.

- s : scalar or array_like, shape (n, ), optional
The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.

- c : color, sequence, or sequence of color, optional
The marker color. Possible values:

   * A single color format string.
   * A sequence of color specifications of length n.
   * A sequence of n numbers to be mapped to colors using cmap and norm.
   * A 2-D array in which the rows are RGB or RGBA.
Note that c should not be a single numeric RGB or RGBA sequence because that is indistinguishable from an array of values to be colormapped. If you want to specify the same RGB or RGBA value for all points, use a 2-D array with a single row. Otherwise, value- matching will have precedence in case of a size matching with x and y.

marker : MarkerStyle, optional
The marker style. marker can be either an instance of the class or the text shorthand for a particular marker. Defaults to None, in which case it takes the value of rcParams["scatter.marker"] = 'o' = 'o'. See markers for more information about marker styles.

cmap : Colormap, optional, default: None
A Colormap instance or registered colormap name. cmap is only used if c is an array of floats. If None, defaults to rc image.cmap.

```python
# Scatter Plot
x = np.random.rand(50) # returns 50 random values between 0 and 1
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='inferno')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar() # show color scale
plt.grid(True)
plt.show()
```

#### Bar Plot

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

#### Subplots



```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sqrt(x)

fig, axs = plt.subplots(2, 1, figsize=(10, 10))
# plt.subplots(2, 1) returns a tuple with 2 axes objects
# axs[0] is the first axes object
# axs[1] is the second axes object
# in plt.subplots(2, 1), 2 is the number of rows and 1 is the number of columns

axs[0].plot(x, y1, 'r')
axs[0].set_title('Sine Wave')
axs[1].plot(x, y2, 'b')
axs[1].set_title('Cosine Wave')

for ax in axs:
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

        fig.add_subplot((nrows, ncols, index))
- index means the position of the subplot in the grid
- 111 means 1 row, 1 column, and the first subplot
- projection='3d' means 3D plot

```python
fig = plt.figure(figsize=(15, 6))
fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(2, 3, 1, projection='3d')
# 2, 3, 1 means 2 rows, 3 columns, and the first plot
plt.show()

```
