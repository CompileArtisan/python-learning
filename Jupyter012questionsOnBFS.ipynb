4. Find the minimum number of throws required to win a given Snakes and Ladders board game.
For example, the following game requires at least 7 dice throws to win: <br><br>
 


<br><br>
We construct a directed graph, keeping in mind the following conditions:
1.	For any vertex in graph v, we have an edge from v to v+1, v+2, v+3, v+4, v+5, v+6 as we can reach any of these nodes in one throw of dice from node v.
2.	If any of these neighbors of v has a ladder or snake, which takes us to position x, then x becomes the neighbor instead of the base of the ladder or head of the snake.
Solve using BFS


### Solution

![WhatsApp Image 2024-08-14 at 22.59.52_db9227c1.jpg](<attachment:WhatsApp Image 2024-08-14 at 22.59.52_db9227c1.jpg>)

<br><br>

### Code

```python
board = [[] for i in range(101)]
for i in range(101):
    for j in range(1, 7):
        if i+j < 101:
            board[i].append(i+j)



snakes = {17: 7, 54: 34, 62: 19, 64:60, 87:36, 93:73, 95:75, 98: 79}
ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 51: 67, 72: 91, 80: 99}

for i in board:
    for j in i:
        if j in ladders:
            i[i.index(j)] = ladders[j]
        if j in snakes:
            i[i.index(j)] = snakes[j]

def min_dice_throws(board):
    visited = [False]*101
    queue = [(0, 0, [])]
    visited[0] = True
    while queue:
        cell, dice, path = queue.pop(0)
        if cell == 100:
            print("Path taken:", ' -> '.join(map(str, path + [100])))
            return dice
        for i in board[cell]:
            if not visited[i]:
                visited[i] = True
                new_path = path + [cell]
                queue.append((i, dice+1, new_path))
    return -1

print("The minimum number of dice throws is:", min_dice_throws(board))
```
