1. In the given problem ,A is the start node and F is the Goal Node <br><br>


```python
from Python029graphs003adjacencyList import Graph
from Python032graphs006breadthFirstSearch import breadthFirstSearch

graph = Graph(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I','J', 'K', 'L', 'M'])
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'D')
graph.add_edge('B', 'E')
graph.add_edge('C', 'F')
graph.add_edge('C', 'G')
graph.add_edge('D', 'H')
graph.add_edge('D', 'I')
graph.add_edge('E', 'J')
graph.add_edge('G', 'K')
graph.add_edge('H', 'L')
graph.add_edge('J', 'M')
bfs=breadthFirstSearch(graph.graph, 'A')
for i in bfs:
    if i == 'F':
        print(f"(({i}))", end=' ')
    else:
        print(i, end=' ')
```
