### np.array()


- object: This parameter represents the input data that you want to convert into an array. It can be an array-like object, an object exposing the array interface, or a nested sequence. If object is a scalar, a 0-dimensional array containing that scalar value will be returned.

- dtype: This parameter specifies the desired data type for the array. If not provided, NumPy will try to use a default data type that can represent the values in the input data.

- copy: This parameter determines whether a copy of the input data should be made or not. If copy is set to True (default), then the object is copied. If copy is set to False, a copy will only be made if necessary, such as when the input data is a nested sequence or if a copy is needed to satisfy other requirements like dtype or order.

- order: This parameter specifies the memory layout of the array. If the input data is not already an array, the newly created array will be in C order (row major) by default, unless 'F' is specified, in which case it will be in Fortran order (column major). If the input data is already an array, the order parameter has different effects depending on the value of copy. If copy is False, the order of the input array will be preserved unless 'F' is specified, in which case the array will be converted to Fortran order. If copy is True, the resulting array will have the most similar order to the input array, unless 'F' is specified, in which case the resulting array will be in Fortran order.

- subok: This parameter determines whether sub-classes will be passed through or not. If subok is True, then sub-classes will be preserved in the resulting array. If subok is False (default), the resulting array will be forced to be a base-class array.

- ndmin: This parameter specifies the minimum number of dimensions that the resulting array should have. If the input data has fewer dimensions than ndmin, ones will be prepended to the shape of the array to meet this requirement.

- like: This parameter allows the creation of arrays that are not NumPy arrays, but are compatible with the object passed in via this argument. If the like object supports the __array_function__ protocol, the result will be defined by it.







````python
print(np.array)

# List to array
np.array([1, 2, 3])  # [1 2 3]

# Nested list to 2D array
np.array([[1, 2], [3, 4]])  # [[1 2]
                            #  [3 4]]

# Single value to 0D array
np.array(5)  # 5
# Integer array
np.array([1, 2, 3], dtype=int)  # [1 2 3]

# Float array
np.array([1, 2, 3], dtype=float)  # [1. 2. 3.]

# String array
np.array([1, 2, 3], dtype=str)  # ['1' '2' '3']


x = np.array([1, 2, 3])
a = np.array(x, copy=True)
b = np.array(x, copy=False)

x[0] = 99
print(a)  # [1 2 3]  (not affected)
print(b)  # [99 2 3] (affected)
original = [1, 2, 3]
a = np.array(x, copy=True)
b = np.array(x, copy=False)

x[0] = 99
print(a)  # [1 2 3]  (not affected)
print(b)  # [99 2 3] (affected)
class MyArray(np.ndarray):
    pass

a = MyArray([1, 2, 3])
b = np.array(a, subok=True)   # Uses MyArray
c = np.array(a, subok=False)  # Uses standard ndarray
# 1D array
np.array([1, 2, 3], ndmin=1)  # [1 2 3]

# 2D array
np.array([1, 2, 3], ndmin=2)  # [[1 2 3]]
````

### array.reshape()

- Reshape to a different shape: You can use ```array.reshape()``` to change the shape of an array to a specified shape. For example, if you have a 1D array with 12 elements, you can reshape it into a 2D array with 3 rows and 4 columns using ```array.reshape(3, 4)```.

- Reshape to a different number of dimensions: ```array.reshape()``` can also be used to change the number of dimensions of an array. For example, you can reshape a 1D array with 12 elements into a 3D array with shape (2, 2, 3) using array.reshape(2, 2, 3).

- Reshape with automatic dimension inference: If you provide only one argument to ```array.reshape()```, it will automatically infer the remaining dimensions based on the total number of elements in the array. For example, if you have a 1D array with 12 elements, you can reshape it into a 3D array with shape (2, 2, 3) using ```array.reshape(2, 2)```.

- Reshape with -1 as a dimension: You can use -1 as one of the dimensions in ```array.reshape()``` to let NumPy automatically calculate the size of that dimension based on the total number of elements and the other specified dimensions. For example, if you have a 1D array with 12 elements, you can reshape it into a 2D array with shape (3, -1) using ```array.reshape(3, -1)```, and NumPy will automatically calculate the number of columns as 4.

<br><br>

It's important to note that the total number of elements in the original array must be equal to the total number of elements in the reshaped array. Otherwise, a ValueError will be raised.

#### Plotting the equation y=sin(x)
        np.linspace(start included, end included, number of points)

````python
x = np.linspace(0, 10, 100) # x = [0, 0.1, 0.2, ..., 9.9, 10]
y = np.sin(x) 
plt.figure(figsize=(10, 6)) # width, height in inches
plt.plot(x, y, label='Sine Wave') # label for legend
plt.title('Basic Line Plot') # title
plt.xlabel('X-axis') 
plt.ylabel('Y-axis')
plt.legend() # show legend
plt.grid(True) # show grid
plt.show()
````

- arange() is used when you know start and end points, and <b> step value </b>
- linspace() is used when you know start and end points, and the <b> number of points </b>

#### Scatter Plots
        plt.scatter(x, y, s (colors), s (sizes), a (alpha), cmap (colormap))
- x, y : array_like, shape (n, )
The data positions.

- s : scalar or array_like, shape (n, ), optional
The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.

- c : color, sequence, or sequence of color, optional
The marker color. Possible values:

   * A single color format string.
   * A sequence of color specifications of length n.
   * A sequence of n numbers to be mapped to colors using cmap and norm.
   * A 2-D array in which the rows are RGB or RGBA.
Note that c should not be a single numeric RGB or RGBA sequence because that is indistinguishable from an array of values to be colormapped. If you want to specify the same RGB or RGBA value for all points, use a 2-D array with a single row. Otherwise, value- matching will have precedence in case of a size matching with x and y.

marker : MarkerStyle, optional
The marker style. marker can be either an instance of the class or the text shorthand for a particular marker. Defaults to None, in which case it takes the value of rcParams["scatter.marker"] = 'o' = 'o'. See markers for more information about marker styles.

cmap : Colormap, optional, default: None
A Colormap instance or registered colormap name. cmap is only used if c is an array of floats. If None, defaults to rc image.cmap.

````python
# Scatter Plot
x = np.random.rand(50) # returns 50 random values between 0 and 1
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='inferno')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar() # show color scale
plt.grid(True)
plt.show()
````

#### Subplots



````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sqrt(x)

fig, axs = plt.subplots(2, 1, figsize=(10, 10))
# plt.subplots(2, 1) returns a tuple with 2 axes objects
# axs[0] is the first axes object
# axs[1] is the second axes object
# in plt.subplots(2, 1), 2 is the number of rows and 1 is the number of columns

axs[0].plot(x, y1, 'r')
axs[0].set_title('Sine Wave')
axs[1].plot(x, y2, 'b')
axs[1].set_title('Cosine Wave')

for ax in axs:
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.grid(True)

plt.tight_layout()
plt.show()
````
